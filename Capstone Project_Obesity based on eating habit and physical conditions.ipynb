It is required to encode the categorical features in order to be able to use KNN method for Machine Learning!
In this set of materials three different methods are used to encode the categorical variable. The reason is the need to manipulation of the codes. 

```python
replace_map={'NObeyesdad':{'Normal_Weight': 0, 'Overweight_Level_I': 1,'Overweight_Level_II': 2, 'Insufficient_Weight':-1, 'Obesity_Type_I' : 3  , 'Obesity_Type_II' : 4, 'Obesity_Type_III' : 5 }}
replace_map2={'CALC':{'no':0, 'Sometimes':1, 'Frequently':2, 'Always':3}}
replace_map3={'CAEC':{'no':0, 'Sometimes':1, 'Frequently':2, 'Always':3}}
replace_map4={'MTRANS':{'Walking':0, 'Public_Transportation':1, 'Automobile':2, 'Motorbike':2, 'Bike':2}}
```

```python
obesitydf.replace(replace_map, inplace=True)
obesitydf.replace(replace_map2, inplace=True)
obesitydf.replace(replace_map3, inplace=True)
obesitydf.replace(replace_map4, inplace=True)
```

## Knn Method

```python
XX=obesitydf_prep.drop(['NObeyesdad'], axis=1)
ss=StandardScaler()
X=ss.fit_transform(XX)
```

```python
y=obesitydf_prep['NObeyesdad']
```

```python
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.30, random_state=42, stratify=y)
```

```python
knn5=KNeighborsClassifier(n_neighbors=5, p=2, metric='minkowski', weights='distance', n_jobs=-1)
```

```python
u=Knn5.fit(X_train, y_train)
y_pred = u.predict(X_test)
```

```python
# Estimating the accuracy of the model!

skf=StratifiedKFold(n_splits=5)
cv_indices=skf.split(X,y)
cv_indices=[[tr,te] for tr, te in cv_indices]
def accuracy_crossvalidator(X, y, knn, cv_indices):
    scores=[]
    for train_i, test_i in cv_indices:
        X_train = X[train_i, :]
        X_test = X[test_i, :]
        Y_train = y[train_i]
        Y_test = y[test_i]
        knn.fit(X_train, Y_train)
        acc = knn.score(X_test, Y_test)
        scores.append(acc)
        print(('Fold accuracy:', acc))
    print(('Mean CV accuracy:', np.mean(scores)))
    return 
knn5=KNeighborsClassifier(n_neighbors=5, p=2, metric='minkowski', weights='distance', n_jobs=-1)
accuracy_crossvalidator(X, y, knn5, cv_indices)
```

## The model predicts correctly in 82% of cases, in average!
